```python
# ==========================================
# Project 3 : AI Recommendation Logic
# DecodeLabs AI Internship
# ==========================================

print("=" * 55)
print("      AI RECOMMENDATION SYSTEM")
print("=" * 55)

# Dataset
recommendations = {
    "movies": [
        "Inception",
        "Interstellar",
        "The Dark Knight",
        "Avatar",
        "The Matrix"
    ],
    "books": [
        "Atomic Habits",
        "Rich Dad Poor Dad",
        "The Alchemist",
        "Think and Grow Rich",
        "Harry Potter"
    ],
    "music": [
        "Shape of You",
        "Blinding Lights",
        "Believer",
        "Perfect",
        "Faded"
    ],
    "sports": [
        "Football",
        "Cricket",
        "Basketball",
        "Tennis",
        "Badminton"
    ],
    "technology": [
        "Artificial Intelligence",
        "Machine Learning",
        "Python Programming",
        "Cyber Security",
        "Cloud Computing"
    ]
}

print("\nAvailable Categories:")
for category in recommendations:
    print("-", category.title())

print()

# User Input
choice = input("Enter your favorite category: ").lower()

print()

# Recommendation Logic
if choice in recommendations:
    print("Recommended Items:\n")

    for item in recommendations[choice]:
        print("•", item)

else:
    print("Sorry!")
    print("No recommendations available for this category.")

print("\nThank you for using the AI Recommendation System!")
```
